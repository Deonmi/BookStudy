# 4.4 简易散点图

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
import numpy as np
```

## 4.4.1 用plt.plot画散点图

```python
x = np.linspace(0, 10, 30)
```

```python
y = np.sin(x)
```

```python
plt.plot(x, y, 'o', color='black')
```

图形符号类型

```python
rng = np.random.RandomState(0)
```

```python
for marker in ['o', '.', ',', 'x', '+', 'v', '^', '<', '>', 's', 'd']:
    plt.plot(rng.rand(5), rng.rand(5), marker, label="marker='{0}'".format(marker))
    plt.legend(numpoints=1)
    plt.xlim(0, 1.8)
```

与线条、颜色代码组合

```python
plt.rcParams['font.sans-serif'] = ['FangSong'] # 指定默认字体
plt.rcParams['axes.unicode_minus'] = False # 解决保存图像是负号'-'显示为方块的问题
plt.plot(x, y, '-ok')
plt.title('直线，圆圈，黑色')
```

plt.plot支持许多设置线条和散点属性的参数。

```python
plt.plot(x, y, '-p', color='gray',
        markersize=15, linewidth=4,
        markerfacecolor='white',
        markeredgecolor='gray',
        markeredgewidth=2)
```

Python – numpy.random.randn()与rand()的区别：
- numpy.random.randn(d0,d1,…,dn)是从标准正态分布中返回一个或多个样本值。
- numpy.random.rand(d0,d1,…,dn)的随机样本位于[0,1)中。

```python
x = rng.randn(100)
y = rng.randn(100)
colors = rng.rand(100)
sizes = 1000 * rng.rand(100)
```

```python
plt.scatter(x, y, c=colors, s=sizes, alpha=0.3, cmap='viridis')
plt.colorbar()
```

## 4.5 可视化异常处理

## 4.5.1 基本误差线

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
import numpy as np
```

```python
x = np.linspace(0, 10, 50)
dy = 0.8
y = np.sin(x) + dy * np.random.randn(50)

plt.errorbar(x, y, yerr=dy, fmt='.k')
```

误差线图形的绘图风格：让误差线的颜色比数据点的颜色浅一点。

```python
plt.errorbar(x, y, yerr=dy, fmt='o', color='black', ecolor='lightgray', elinewidth=3, capsize=0)
```

## 4.5.2 连续误差

```python
from sklearn.gaussian_process import GaussianProcessRegressor
```

```python
# 定义模型和要画的数据
model = lambda x: x * np.sin(x)
xdata = np.array([1, 3, 5, 6, 8])
ydata = model(xdata)

# 计算高斯过程拟合结果
gp = GaussianProcessRegressor(corr='cubic', theta0=1e-2, thetaL=1e-4, thetaU=1E-1, random_state=100)
gp.fit(xdata[:, np.newaxis], ydata)

xfit = np.linspace(0, 10, 1000)
yfit, MSE = gp.predic(xfit[:, np.newaxis], eval_MSE=True)
dyfit = 2 * np.sqrt(MSE)
```
