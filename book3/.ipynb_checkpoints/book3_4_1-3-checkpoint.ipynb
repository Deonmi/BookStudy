```python
plt.style.use('ggplot')
```

## 4.1.3 用不用show()?如何显示图形

1. 在脚本中画图

```python
import numpy as np
```

```python
x = np.linspace(0, 10, 100)
```

```python
plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))
plt.show()
```

2. 在IPython shell中画图

```python
%matplotlib inline
```

3. 在IPython Notebook中画图

```python
x = np.linspace(0, 10, 100)
```

```python
fig = plt.figure()
plt.plot(x, np.sin(x), '-')
plt.plot(x, np.cos(x), '--')
fig.savefig('D:/Python3Space/BookStudy/pic/book3/my_figure.png')
```

## 4.1.4 将图形保存为文件
